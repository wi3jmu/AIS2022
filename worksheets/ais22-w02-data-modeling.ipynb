## 1. Entity-Relationship Model<a id="em_model"></a>

***Recap***

*From [wikipedia]()*: An entityâ€“relationship model (or ER model) describes interrelated things of interest in a specific domain of knowledge. A basic ER model is composed of entity types (which classify the things of interest) and specifies relationships that can exist between entities (instances of those entity types).

The main components of an ER Model are:
- Entity: Entirety of similar objects
- Relationship: Entirety of relationships between object
- Attribute: Property of an entity or relationship type
    - Key attributes uniquely identify an entity
- Cardinalities: The number of times an entity of an entity set participates in a relationship
